```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
p_b = pd.read_csv('Exp_base_/Pix_base/generated_images/ep_200/im_px_stats.csv')
p_r = pd.read_csv('Exp_base_/Pix_Mask_R_Exp/generated_images/ep_200/im_px_stats.csv')
p_m = pd.read_csv('Exp_base_/Pix_Mask_R_Exp_M/generated_images/ep_200/im_px_stats.csv')
p_c = pd.read_csv('Exp_base_/Pix_Mask_R_Exp_B/generated_images/ep_200/im_px_stats.csv')

u_b = pd.read_csv('Exp_base_/UNet_base/generated_images/ep_200/im_px_stats.csv')
u_r = pd.read_csv('Exp_base_/UNet_Mask_R_Exp/generated_images/ep_200/im_px_stats.csv')
u_m = pd.read_csv('Exp_base_/UNet_Mask_R_Exp_M/generated_images/ep_200/im_px_stats.csv')
u_c = pd.read_csv('Exp_base_/UNet_Mask_R_Exp_B/generated_images/ep_200/im_px_stats.csv')


df = {'MAE' : np.array([np.mean(u_b['mae_fi']), np.mean(u_r['mae_fi']), np.mean(u_m['mae_fi']), np.mean(u_c['mae_fi']), \
                        np.mean(p_b['mae_fi']), np.mean(p_r['mae_fi']), np.mean(p_m['mae_fi']), np.mean(p_c['mae_fi'])]) , 
      'PSNR': np.array([np.mean(u_b['psnr_fi']), np.mean(u_r['psnr_fi']), np.mean(u_m['psnr_fi']), np.mean(u_c['psnr_fi']), \
                        np.mean(p_b['psnr_fi']), np.mean(p_r['psnr_fi']), np.mean(p_m['psnr_fi']), np.mean(p_c['psnr_fi'])]) , 
      'LPIPS': np.array([np.mean(u_b['lpips_fi']), np.mean(u_r['lpips_fi']), np.mean(u_m['lpips_fi']), np.mean(u_c['lpips_fi']), \
                        np.mean(p_b['lpips_fi']), np.mean(p_r['lpips_fi']), np.mean(p_m['lpips_fi']), np.mean(p_c['lpips_fi'])]), 
      'SSIM': np.array([np.mean(u_b['ssim_fi']), np.mean(u_r['ssim_fi']), np.mean(u_m['ssim_fi']), np.mean(u_c['ssim_fi']), \
                        np.mean(p_b['ssim_fi']), np.mean(p_r['ssim_fi']), np.mean(p_m['ssim_fi']), np.mean(p_c['ssim_fi'])]), 
      'Configuracion': np.array(['base', 'R', 'M', 'B', \
                         'base', 'R', 'M', 'B',]), 
      'Tipo Modelo': np.array(['UNet', 'UNet', 'UNet', 'UNet', \
                       'Pix2Pix', 'Pix2Pix', 'Pix2Pix', 'Pix2Pix'])}

df = pd.DataFrame.from_dict(df)

plt.style.use('default')

_, (ax, ax2) = plt.subplots (1,2, figsize=(10,5))
sns.scatterplot(data=df, x="MAE", y="PSNR", style='Configuracion', hue='Tipo Modelo', s=100, ax=ax) #, legend=None
# sns.scatterplot(data=df, x="MAE", y="PSNR", style='Model', c='m', s=50, ax=ax)


sns.scatterplot(data=df, x="MAE", y="SSIM", style='Configuracion', hue='Tipo Modelo', s=100, ax=ax2)
# sns.scatterplot(data=df, x="MAE", y="SSIM", style='Model', c='m', s=50, ax=ax2)

# ax.set_ylim(df['F1'].min()-0.02, df['F1'].max()+0.1)
# ax.set_xticks([23.5, 24, 24.5, 25])
# plt.grid(linestyle="--", alpha=0.5)  #just add this
ax.grid (linestyle="--", alpha=0.5)
ax2.grid(linestyle="--", alpha=0.5)
# ax.get_legend().remove()
plt.tight_layout()
plt.savefig("output.png", dpi=200)
```
